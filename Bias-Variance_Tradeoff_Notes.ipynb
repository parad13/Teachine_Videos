## Bias - Varience Tradeoff

**Bias- Varience Tradeoff**
- What is Bias ?
- What is Varience ?
- What is Bias-Varience tradeoff, how to decide which model to use from it ?

- Linear regression


- Graph I :
- degree of polynomial = 1 (linear regression)
- Underfitting - Giving low accuracy on training data, and also low accuracy on validation data
- High bias, high varience condition

- Graph II:
- degree of polynomial = 2 (Quadratic regression)
- Just right(miss fitting good on training data) - Giving good accuracy for training data as well as validation data
- Low bias, Low varience condition

- Graph III :
- degree of polynomial = more than 2 (Multinomial regression)
- Overfitting - Giving high accuracy for training data, but low accuracy on validation data
- Low bias, High varience condition

- High bias - Simple model for our data
- High varience - Complex model for our data


- Bias, Varience, model simplicity/complexity relation



**Key Take away points :**
- In simple words we cany say bias is error on the `training data` & varience is error on `validation data`
- **Bias** - Error on training data
- **Varience** - Error on validation data
- **Overfitting** : High varience means our model only fitting on our training data, Understanding only same type of data & won't be able to interpret on new data(validation data)
- **Underfitting** : High bias means not able to understand(Find any patterns in data) training data as well as validation data
- **Good fit** : Low bias, low varience, our model able to understand training dataas well as testing data **which is our main goal**

```python
#Import important libraries which we will require 
# while performing this task
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from mlxtend.evaluate import bias_variance_decomp
```

```python
#Loading dataset using pandas read_csv() function
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/housing.csv"
df = pd.read_csv(url, header=None)
df.head() # will print first 5 values of the data
```

```python
# Separate dataset into input & output 
data = df.values
X, y = data[:, :-1], data[:, -1] #output -> prdiction, input -> input
# input = [4, 16, 25, 36, 84] -> X
# output=[2, 4, 5, 6, 8] -> y
# Y = X^2 
# model_predict([1, 7, 9]) -> [1, 49 ,81]
```

```python
#split the data into training set & validation set using sklearns train_test_split module
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
#traning size = 70% of data, teststing data size = 30%
```

```python
# define the model
model = LinearRegression() #creating an instance of linear regression
model.fit(X_train, y_train) 
```

```python
model.predict(X_test)
```

```python
14.5 - 13.51882716 #predicted value - actual value = error
```

```python
y_test
```

```python
# estimate bias & variance
mse, bias, var = bias_variance_decomp(model, X_train, y_train, X_test, y_test, loss='mse', num_rounds=200, random_seed=1)
```

```python
# Result evaluation
print('MSE: %.3f' % mse)
print('Bias: %3f' %bias)
print('Variance: %3f' %var)
#best model -> low bias, low varience, low error(mse)
```

<p style = "color:blue">In this case, we can see that the model has a high bias and a low variance. This is to be expected given that we are using a linear regression model. We can also see that the sum of the estimated mean and variance equals the estimated error of the model, e.g. 20.726 + 1.761 = 22.487.<p>
